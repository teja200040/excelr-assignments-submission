```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
data = pd.read_csv('wine (1).csv')
data
```

```python
data.info()
```

```python
#normalizing data
data_normal= scale(data)
```

```python
data_normal
```

```python
pca = PCA(n_components=14)
pca_values = pca.fit_transform(data_normal)
pca_values
```

```python
#loading weights
pca.components_
```

```python
#the amount of variance that each pca explain is
var = pca.explained_variance_ratio_
var
```

```python
#cumulative variance
var1 = np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
#variance plot for pca components obtained
plt.plot(var1,color = "red")
```

```python
pca_values[:,0:1]
```

```python
final = pd.concat([pd.DataFrame(pca_values[:,0:2],columns=['pc1','pc2']), data[['Type']]], axis = 1)
final
```

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
import seaborn as sns
sns.scatterplot(data=final,x='pc1',y='pc2',hue='Type',s = 100)
```

```python
p1 = sns.scatterplot(data=final,x='pc1',y='pc2',s = 100)  
for line in range(0,final.shape[0]):
     p1.text(final.pc1[line], final.pc2[line], final.Type[line], horizontalalignment='left', size='medium')
        
```

# Hclustering

```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt 
```

```python
df = pd.read_csv('wine (1).csv')
df
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)  
```

```python
df_norm = norm_func(df.iloc[:,1:])
df_norm
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch 
z = linkage(df_norm, method="average",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    
)
plt.show()     
```

```python
from sklearn.cluster import AgglomerativeClustering 
import warnings 
warnings.filterwarnings('ignore')
h_complete = AgglomerativeClustering(n_clusters=5, linkage='average',affinity = "euclidean").fit(df_norm) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
df['clust']=cluster_labels # creating a  new column and assigning it to new column 
df
```

```python
df.iloc[:,1:].groupby(df.clust).mean()   
```

```python
data = df[(df.clust==1)]
data   
```

# Kmeans

```python
import pandas as pd
import matplotlib.pylab as plt
from sklearn.cluster import KMeans
import numpy as np
```

```python
df = pd.read_csv('wine (1).csv')
df
```

```python
def norm_func(i):
    x = (i-i.min()) / (i.max() - i.min())
    return (x)  
```

```python
df_norm = norm_func(df.iloc[:,1:]) 
df_norm
```

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()  
```

```python
clf = KMeans(n_clusters=5)
y_kmeans = clf.fit_predict(df_norm)  
```

```python
y_kmeans
```

```python
clf.cluster_centers_
```

```python
clf.inertia_
```

```python
md = pd.Series(y_kmeans)
df['clust'] = md
df
```

```python
df.iloc[:,1:4].groupby(df.clust).mean() 
```

```python
df.plot(x="Alcohol",y ="Malic",c=clf.labels_,kind="scatter",s=30 ,cmap=plt.cm.coolwarm) 
plt.title('Clusters using KMeans')   
```
