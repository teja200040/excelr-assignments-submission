# KMeans clustering

```python
import pandas as pd 
import numpy as np
from sklearn.cluster import KMeans
```

```python
yt = pd.read_excel('EastWestAirlines.xlsx',sheet_name ='data')
yt
```

```python
trans = MinMaxScaler()
trans_df = trans.fit_transform(df)
trans_df
```

```python
fig = plt.figure(figsize=(10, 8))
WCSS = []    
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(trans_df)
    WCSS.append(clf.inertia_) # inertia another name for wcss
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()  
```

```python
# in above there is 6 clusters
```

```python
clf = KMeans(n_clusters=6)
y_Kmeans = clf.fit_predict(trans_df)
```

```python
y_Kmeans
clf.labels_
```

```python
clf.cluster_centers_
```

```python
clf.inertia_
```

```python
md = pd.Series(y_Kmeans)
yt['clust'] = md
yt
```

```python
yt.groupby(yt.clust).mean()
```

```python
yt.plot(x="Balance",y = "Bonus_miles",c=clf.labels_,kind = "scatter",s = 30,cmap=plt.cm.coolwarm_r)
plt.title('Clusters using kmeans')
```

# DB Scan

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
pt = pd.read_excel('EastWestAirlines.xlsx',sheet_name='data')
pt
```

```python
array = pt.values
array
```

```python
stscalar = StandardScaler().fit(array)
X = stscalar.transform(array)
X
```

```python
dbscan = DBSCAN(eps = 0.475,min_samples=5)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
#here -1 implies noise
```

```python
c1 = pd.DataFrame(dbscan.labels_,columns=['cluster'])
c1
```

```python
pt = pd.concat([pt,c1],axis = 1)
pt
```

```python
import matplotlib.pyplot as plt 
pt.plot(x="Balance",y ="Bonus_Miles",c=dbscan.labels_ ,kind="scatter",s=100 ,cmap=plt.cm.copper) 
plt.title('Clusters using DBScan')   
```
