```python
df_train1 = df_train.iloc[:,0:13]

df_train1 = pd.get_dummies(df_train1)
df_train1
```

```python
df_test1 =df_test.iloc[:,0:13]

df_test1 = pd.get_dummies(df_test1)
df_test1
```

```python
#since, there is more columns we have to use PCA
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
sc.fit(df_train1)
train_norm = sc.transform(df_train1)
train_norm
```

```python
sc.fit(df_test1)
test_norm = sc.transform(df_test1)
test_norm 
```

```python
from sklearn.decomposition import PCA
train_pca = PCA(n_components = 102)
train_pca_values = train_pca.fit_transform(train_norm)
train_pca_values
```

```python
test_pca = PCA(n_components = 102)
test_pca_values = test_pca.fit_transform(test_norm)
test_pca_values
```

```python
Var = train_pca.explained_variance_ratio_
Var 
```

```python
# Cumulative variance 
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
plt.plot(var1,color="red")
```

```python
# Selecting first 90 PCAs out of total 102
finaltrain = pd.concat([pd.DataFrame(train_pca_values[:,0:90]),df_train[['Salary']]], axis = 1)
finaltrain
```

```python
finaltest = pd.concat([pd.DataFrame(test_pca_values[:,0:90]),df_test[['Salary']]], axis = 1)
finaltest
```

```python
#Since the training dataset is huge, we'll use some part of it for the training purpose, to reduce time consumed.
array = finaltrain.values
X = array[0:1000,0:90]
Y = array[0:1000,90]
```

# SVM Classification

by using Grid search CV to find best value for params

```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[0.9,0.5,0.1],'C':[1,10,100] },
             {'kernel':['linear'],'C':[1,10,100]}]
gsc = GridSearchCV(clf,param_grid,cv=10,n_jobs=-1)
gsc.fit(X,Y)
```

```python
gsc.best_params_ , gsc.best_score_
```

```python
#Since the testing dataset is huge, we'll use some part of it for the test purpose, to reduce the time consumed.
x = finaltest.values[0:1000,0:90]
y = finaltest.values[0:1000,90]
```

```python
#SVM Clasification
clf = SVC(C=10, kernel='linear')
clf.fit(x,y)
results = clf.score(x,y)
print(np.round(results, 4))
```
